```python
correct_price_col = "Price ($)"

# Check unique values in facility columns
print(location_1_data[["Has Garage", "Has Garden", "Has Swimming Pool"]].head())
print(location_1_data[["Has Garage", "Has Garden", "Has Swimming Pool"]].dtypes)

# Convert facility columns to categorical if needed
for col in ["Has Garage", "Has Garden", "Has Swimming Pool"]:
    if location_1_data[col].dtype == 'bool':  # If values are True/False
        location_1_data[col] = location_1_data[col].map({True: "Yes", False: "No"})
    elif location_1_data[col].dtype in ['int64', 'float64']:  # If values are 1/0
        location_1_data[col] = location_1_data[col].map({1: "Yes", 0: "No"})

# Melt data
facility_data = location_1_data.melt(id_vars=[correct_price_col], 
                                     value_vars=["Has Garage", "Has Garden", "Has Swimming Pool"], 
                                     var_name="Facility", value_name="Has Facility")

# Print data to confirm changes
print(facility_data.head())

# Check if the data is empty before plotting
if facility_data.empty:
    print("Error: facility_data is empty. Check column names and values.")
else:
    # Plot boxplot
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Facility", y=correct_price_col, hue="Has Facility", data=facility_data, palette="viridis")

    # Labels
    plt.xlabel("Facility Type")
    plt.ylabel("Property Price ($)")
    plt.title("Impact of Facilities on Property Prices in Location 1")

    # Show plot
    plt.show()
```

```python
# How does proximity to schools, hospitals, or shopping malls affect demand?
location_1 = df[df["Location"] == "Location 1"]

# Convert proximity to categorical (Near/Far based on median)
for facility in ["Nearby Schools", "Nearby Hospitals", "Nearby Shopping Malls"]:
    median_distance = location_1[facility].median()
    location_1[facility] = location_1[facility].apply(lambda x: "Near" if x <= median_distance else "Far")

# Melt data for visualization
demand_data = location_1.melt(id_vars=["Price ($)"], value_vars=["Nearby Schools", "Nearby Hospitals", "Nearby Shopping Malls"], 
                              var_name="Facility", value_name="Proximity")

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Facility", y="Price ($)", hue="Proximity", data=demand_data, palette="coolwarm")

# Labels
plt.xlabel("Facility Type")
plt.ylabel("Property Price ($)")
plt.title("Impact of Proximity to Schools, Hospitals, and Shopping Malls on Property Demand")

plt.legend(title="Proximity", loc="upper left")
plt.show()
```

```python
price_column = "Price ($)"

# Check if required columns exist
required_columns = ["Location", price_column, "Region", "Has Garage", "Has Garden", "Has Swimming Pool"]
missing_cols = [col for col in required_columns if col not in df.columns]
if missing_cols:
    raise KeyError(f"Missing columns in dataset: {missing_cols}")

# Convert facility columns to numeric
facility_columns = ["Has Garage", "Has Garden", "Has Swimming Pool"]
df[facility_columns] = df[facility_columns].apply(pd.to_numeric, errors="coerce").fillna(0)

# Filter properties in Central region
central_df = df[df["Region"] == "Central"]

# Ensure data exists for "Central"
if central_df.empty:
    raise ValueError("No data found for Central region. Check if 'Region' column values are correct.")

# Group by location and calculate average price
top_central_locations = central_df.groupby("Location").agg(avg_price=(price_column, "mean"))

# Sum up facilities
facility_sums = central_df.groupby("Location")[facility_columns].sum()

# Merge price and facility data
top_central_locations = top_central_locations.merge(facility_sums, left_index=True, right_index=True)

# Calculate total facility score
top_central_locations["facility_score"] = top_central_locations[facility_columns].sum(axis=1)

# Sort by price and facilities
top_central_locations = top_central_locations.sort_values(by=["avg_price", "facility_score"], ascending=[False, False])

# Select top 5 locations
top_5 = top_central_locations.head(5)

# Plot graph
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for average price
sns.barplot(x=top_5.index, y=top_5["avg_price"], ax=ax1, color="#22b0cf", edgecolor="black")
ax1.set_ylabel("Average Price ($)", color="black")
ax1.set_xlabel("Location")
ax1.set_title("Top 5 Locations in Central Region by Price & Facilities")
ax1.tick_params(axis="y", labelcolor="black")

# Line plot for facility count
ax2 = ax1.twinx()
ax2.plot(top_5.index, top_5["facility_score"], color="red", marker="o", linestyle="-", linewidth=2, label="Facility Score")
ax2.set_ylabel("Total Facilities", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Show legend
ax2.legend(loc="upper right")

plt.xticks(rotation=45)
plt.show()
```

```python
# Define correct column names
price_column = "Price ($)"
facility_columns = ["Has Garage", "Has Garden", "Has Swimming Pool"]

# Check if required columns exist
required_columns = ["Location", price_column, "Region"] + facility_columns
missing_cols = [col for col in required_columns if col not in df.columns]
if missing_cols:
    raise KeyError(f"Missing columns in dataset: {missing_cols}")

# Convert facility columns to numeric
df[facility_columns] = df[facility_columns].apply(pd.to_numeric, errors="coerce").fillna(0)

# Filter properties in Central region
central_df = df[df["Region"] == "Central"]

# Ensure data exists for "Central"
if central_df.empty:
    raise ValueError("No data found for Central region. Check if 'Region' column values are correct.")

# Group by location and calculate average price
top_central_locations = central_df.groupby("Location").agg(avg_price=(price_column, "mean"))

# Sum up facilities for each location
facility_sums = central_df.groupby("Location")[facility_columns].sum()

# Merge price and facility data
top_central_locations = top_central_locations.merge(facility_sums, left_index=True, right_index=True)

# Convert values to percentages
top_central_locations["Price (%)"] = (top_central_locations["avg_price"] / top_central_locations["avg_price"].sum()) * 100
for facility in facility_columns:
    top_central_locations[f"{facility} (%)"] = (top_central_locations[facility] / top_central_locations[facility].sum()) * 100

# Sort by price percentage
top_central_locations = top_central_locations.sort_values(by=["Price (%)"], ascending=[False])

# Select top 5 locations
top_5 = top_central_locations.head(5)

# Define shades of blue for consistency
price_color = "#003f5c"  # Dark Blue for Price
facility_colors = ["#0077b6", "#48cae4", "#90e0ef"]  # Lighter Blue shades for Facilities

# Plot graph
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for price percentage (Darker Blue)
bars1 = sns.barplot(x=top_5.index, y=top_5["Price (%)"], ax=ax1, color=price_color, edgecolor="black", label="Price (%)")
ax1.set_ylabel("Price (%)", color="black")
ax1.set_xlabel("Location")
ax1.set_title("Top 5 Locations in Central Region: Price & Facility Percentage Comparison")
ax1.tick_params(axis="y", labelcolor="black")

# Add labels on price bars
for bar in bars1.patches:
    ax1.annotate(f"{bar.get_height():.1f}%", 
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha="center", va="bottom", fontsize=10, color="black", fontweight="bold")

# Create a secondary Y-axis for facilities
ax2 = ax1.twinx()

# Stacked bar chart for facilities percentage using Matplotlib
bottom_values = np.zeros(len(top_5))
for i, facility in enumerate(facility_columns):
    bars2 = ax2.bar(top_5.index, top_5[f"{facility} (%)"], bottom=bottom_values, color=facility_colors[i], alpha=0.8, label=f"{facility} (%)")
    
    # Add labels on stacked facility bars
    for bar in bars2:
        height = bar.get_height()
        if height > 0:
            ax2.annotate(f"{height:.1f}%", 
                         (bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2), 
                         ha="center", va="center", fontsize=10, color="white", fontweight="bold")

    bottom_values += top_5[f"{facility} (%)"].values

ax2.set_ylabel("Facilities (%)", color="black")
ax2.tick_params(axis="y", labelcolor="black")

# Create a combined legend
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.xticks(rotation=45)
plt.show()
```
